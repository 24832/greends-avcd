Importar pacotes

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob
import re
```

Simulate var1

```python
var1 = []
random.seed(30) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
levels = ["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland" ]
for _ in range(100): # why is a loop needed?
    var1 += random.sample(levels, 1) # var1.append(random.sample(levels, 1)) would also work
    
print("With list:" , random.sample(levels, 1))
print (var1)
# var1 é uma variavel qualitativa nominal
```

Simulate var2

```python
np.random.seed(30) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
var2 = np.random.uniform(0, 100, 100)

print (var2)
# var2 é uma variavel quantitativa contínua
```

Simulate table1

```python
table1 = pd.DataFrame(var1).value_counts(sort=True)
table1 = table1.rename_axis("landuse")
table1 = table1.reset_index(name="Frequency")

print(table1)
# table1 tem duas variéveis a lanuse (qualitativa nominal) e a frequency (quantitativa discreta)
#Proposta do grafico: diagrama de barras e/ou em forma circular (Pie)
```

Criação de grafico de barras

```python
x_axis = np.array(["Agro-Forestry system", "Managed Forest", "Pasture", "Permanent crops","Shrubland","Irrigated crops", "Urban","Natural Forest"])
y_axis = np.array([19, 14, 13, 12, 12, 11, 11, 8])

plt.barh(x_axis, y_axis, color = "#00FF33", height = 0.4)
plt.xlabel("Frequency")
plt.ylabel("Land Use")
plt.title(" Land Use Land Cover 2023")
plt.show()
```

Simulate table3

```python
np.random.seed(30) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
year = list(range(1970,2021))
temp = np.random.normal(17,2,51)
table3 = pd.DataFrame(list(zip(year, temp)), columns = ["Year", "Temperature"])
print(table3)
#table3 tem duas variaveias quantitativas sendo Year (Discreta) e Temperature (Contínua)
#Proposta do grafico: grafico de linha para serie temporal
```

Criar grafico de serie temporal

```python
sns.set(rc={'figure.figsize': (12, 5)})
sns.lineplot(data=table3, x="Year", y="Temperature", markers=False)
plt.title(" Variação de temperatura 51 anos ")
```

Simulate table4

```python
xx = np.array([16,21])
yy = np.array([300, 1200])
means = [xx.mean(), yy.mean()]  
stds = [xx.std() / 3, yy.std() / 3]
corr = -0.7 # correlation
covs = [[stds[0]**2          , stds[0]*stds[1]*corr], 
        [stds[0]*stds[1]*corr,           stds[1]**2]] # covariance matrix
table4 = pd.DataFrame(np.random.multivariate_normal(means, covs, 100), columns = ["Mean Anual Temperature", "Total Precipitation"])

print(table4)
#table4 tem duas variaveias quantitativas (Mean Anual Temperature and Total Precipitation)
#Proposta do grafico: grafico de linha
```

Criar grafico de linha

```python
sns.lineplot(data=table4, x="Mean Anual Temperature", y="Total Precipitation",markers=True, dashes=True)
```

table5

```python
col1 = pd.Series(list(range(1900,2010,10))).repeat(8)
col2 = ["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland" ]*11
col3 = np.random.uniform(0, 100, 90)
table5 = pd.DataFrame(list(zip(col1, col2, col3)), columns = ["Year", "Landuse", "Cover"])

print(table5)
#table3 tem três variaveias sendo uma qualitativa [Landuse (nominal)] e duas quantitativas  [Year (Discreta) e cover (Contínua)]
#Proposta do grafico: grafico de calor
```

Criação de grafico de calor

```python
table5=table5.pivot('Year','Landuse','Cover')
fig, ax = plt.subplots (figsize=(6, 7))
sns.heatmap(table5, annot=table5.rank(axis="columns"), linewidths = .3)
```
