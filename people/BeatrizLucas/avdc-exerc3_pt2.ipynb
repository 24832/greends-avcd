```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Make random samples with different sizes
sample_size = [10, 50, 100, 150, 200, 250, 300, 500, 1000]
```

```python
df = pd.read_csv('EFIplus_medit.zip',compression='zip', sep=";")
```

```python
mean_temp_ann = []

for i in sample_size:
    
    # select a sample size
    sample = df.sample(n=i)
    
    # mean of temp_ann for the sample
    mean_temp = np.mean(sample["temp_ann"])
    
    # append the mean to the list mean_temp_ann
    mean_temp_ann.append(mean_temp)
```

```python
# Plot temp_ann mean for difeerent sample sizes in a bar plot
sns.barplot(x=np.array(sample_size), y=np.array(mean_temp_ann), color = 'darkblue')

# Add a title and lables to the plot
plt.xlabel("Sample size")
plt.ylabel("Mean temp_ann")
plt.title("temp_ann mean for different sample sizes")

# Show the plot
plt.show()
```
