```python
import pandas as pd
import zipfile
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For showing plots
```

```python
df = pd.read_csv('EFIplus_medit.csv', sep=";")
```

```python
df_ebro = df[df['Catchment_name']=='Ebro']
df_galiza = df[df['Catchment_name']=='Galiza-Norte']
df_minho = df[df['Catchment_name']=='Minho']
df_tejo = df[df['Catchment_name']=='Tejo']
```

## Strip plots

```python
plt.figure(1,figsize=(10,10))
plt.gcf().subplots_adjust(left = 0.2, bottom = 0.2, right = 1.2,
                          top = 0.9, wspace = 0.5, hspace = 0.5)
plt.subplot(2,2,1)
sns.stripplot(data=df_ebro, y='temp_ann').set_title("Ebro")
plt.subplot(2,2,2)
sns.stripplot(data=df_galiza, y='temp_ann').set_title("Galiza-Norte")
plt.subplot(2,2,3)
sns.stripplot(data=df_minho, y='temp_ann').set_title("Minho")
plt.subplot(2,2,4)
sns.stripplot(data=df_tejo, y='temp_ann').set_title("Tejo")

plt.show()
```

## Histogramm 

```python
plt.figure(1,figsize=(10,10))
plt.gcf().subplots_adjust(left = 0.2, bottom = 0.2, right = 1.2,
                          top = 0.9, wspace = 0.5, hspace = 0.5)

plt.subplot(2,2,1)
sns.histplot(
    df_ebro["temp_ann"], 
    kde=True,
    stat="density",
    kde_kws=dict(cut=3),
    alpha=.4,
    edgecolor=(1, 1, 1, 0.4),
).set_title("Ebro")

plt.subplot(2,2,2)
sns.histplot(
    df_galiza["temp_ann"], 
    kde=True,
    stat="density",
    kde_kws=dict(cut=3),
    alpha=.4,
    edgecolor=(1, 1, 1, 0.4),
).set_title("Galiza-Norte")

plt.subplot(2,2,3)
sns.histplot(
    df_minho["temp_ann"], 
    kde=True,
    stat="density",
    kde_kws=dict(cut=3),
    alpha=.4,
    edgecolor=(1, 1, 1, 0.4),
).set_title("Minho")

plt.subplot(2,2,4)
sns.histplot(
    df_tejo["temp_ann"], 
    kde=True,
    stat="density",
    kde_kws=dict(cut=3),
    alpha=.4,
    edgecolor=(1, 1, 1, 0.4),
).set_title("Tejo")

plt.show()
```

## Box plots 

```python
plt.figure(1,figsize=(10,10))
plt.gcf().subplots_adjust(left = 0.2, bottom = 0.2, right = 1.2,
                          top = 0.9, wspace = 0.5, hspace = 0.5)
plt.subplot(2,2,1)
sns.boxplot( y="temp_ann", data=df_ebro).set_title("Ebro")
plt.subplot(2,2,2)
sns.boxplot( y="temp_ann", data=df_galiza).set_title("Galiza-Norte")
plt.subplot(2,2,3)
sns.boxplot( y="temp_ann", data=df_minho).set_title("Minho")
plt.subplot(2,2,4)
sns.boxplot( y="temp_ann", data=df_tejo).set_title("Tejo")

plt.show()
```

## mean temperatures

```python
x=[]
y=[]
for k in range (50,1000,50):
    x=x+[k]
    y=y+[df[1:k]['temp_ann'].mean()]

plt.plot(x,y,marker='o')
plt.show()
```
